```python
#Handling wrong values of x, y, z
print('Number of rows with x = 0 are {}'.format((df.x==0).sum()))
print('Number of rows with x = 0 are {}'.format((df.y==0).sum()))
print('Number of rows with x = 0 are {}'.format((df.z==0).sum()))
```

```python
#Setting wrong values of x, y, z with NaN
df.x = df.x.replace(0,np.NaN)
df.y = df.y.replace(0,np.NaN)
df.z = df.z.replace(0,np.NaN)

print("Number of rows with x = 0 are {}".format((df.x==0).sum()))
print("Number of rows with y = 0 are {}".format((df.y==0).sum()))
print("Number of rows with z = 0 are {}".format((df.z==0).sum()))
```
